```python
print("Elementary imports: ")
import os
import json
import glob
import argparse

print("numpy/scipy imports:")
import numpy as np
from scipy import signal as sg
import scipy.ndimage as ndimage
from scipy.ndimage.filters import maximum_filter
import cv2 as cv


print("PIL imports:")
from PIL import Image

print("matplotlib imports:")
import matplotlib.pyplot as plt
```

```python
img = cv.imread('./data/munster_000023_000019_leftImg8bit.png')
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(gray_img,cmap='gray')
```

```python
image = cv.blur(gray_img,(15,15))
plt.imshow(image,cmap='gray')
plt.show()
```

```python
i_height= 250
i_width = 500
# image = image[::4, ::4]
image=cv.resize(image, (i_width, i_height ))
plt.imshow(image,cmap='gray')
plt.show()
```

```python
filter_kernel = np.array([[-1, -1, -1, -1, -1],
                   [-1,  1,  2,  1, -1],
                   [-1,  2,  4,  2, -1],
                   [-1,  1,  2,  1, -1],
                   [-1, -1, -1, -1, -1]])

plt.imshow(filter_kernel,cmap='gray')

res = ndimage.filters.convolve(image, filter_kernel)
plt.imshow(res,cmap='gray')
```

```python
filter_kernel = np.array([[-3.5, -2.25, -2, -2, -2],
                          [-2.25,  -1, 1/4, 1/2, 1],
                          [-2,  1/4, 1, 1, 1.25],
                          [-2, 1/2, 1, 2, 3],
                          [-2, 1,  1.25, 3, 4]])
filter_kernel = np.hstack([filter_kernel[::], np.fliplr(filter_kernel)])
filter_kernel = np.vstack([filter_kernel[::], filter_kernel[::-1]])

plt.imshow(filter_kernel,cmap='gray')

res = ndimage.filters.convolve(image, filter_kernel)
plt.imshow(res,cmap='gray')
```

```python
image = cv.imread('./data/munster_000023_000019_leftImg8bit.png', 0)
image = cv.equalizeHist(image)
kernel = np.ones((5, 5),np.uint8)
image = cv.morphologyEx(image, cv.MORPH_OPEN, kernel, iterations=2)
image = cv.dilate(image,kernel,iterations = 2)

image = cv.GaussianBlur(image, (13, 13), 10)

ret,thresh1 = cv.threshold(image, 100, 255, cv.THRESH_TOZERO)


filter_kernel = np.array([[-3.5, -2.25, -2, -2, -2],
                          [-2.25,  -1, 1/4, 1/2, 1],
                          [-2,  1/4, 1, 1, 1.25],
                          [-2, 1/2, 1, 2, 3],
                          [-2, 1,  1.25, 3, 4]])
filter_kernel = np.hstack([filter_kernel[::], np.fliplr(filter_kernel)])
filter_kernel = np.vstack([filter_kernel[::], filter_kernel[::-1]])
# cv.imwrite(SAVE_PATH, thresh1)
res = ndimage.filters.convolve(thresh1, filter_kernel)
plt.imshow(thresh1,cmap='gray')
```
